# Content Based Filtering

Content-based filtering, also referred to as cognitive filtering, recommends items based on a comparison between similarity content of the items.

In this case I used a dataset of movies that I have used in collaborative filtering. In this Case these films are viewed by their characteristics based on type (Indonesia or Hollywood), Genre, IMDB Rate, and duration. The Output of content based filtering is films that have similarities to the film.

dataset : "dataset film.csv"

### Import Libraries and Dataset

```python
#Import Libraries
import pandas as pd
```

```python
#Read dataset
dataset=pd.read_csv('dataset film.csv')
dataset
```

### 1. Data Preprocessing

### Set Index

```python
#Set index of Movie
datafilm_index=dataset.set_index('Movie')
```

Set Index of 'Movie' because i want to analysis of features that important and get dummies of it in the next step.

```python
datafilm_index
```

### Get Dummies of dataset that convert categorical to numeric

```python
# Get Dummies of data because any categorical feature in dataset and for analysis categorical must be numerical
datafilm_index=pd.get_dummies(datafilm_index)
```

```python
datafilm_index
```

### 2. Analysis

In this set of analysis will be produced movies that have the same characteristics as the movie we are looking for. The steps are:
- Step 1: Defining cosine simalirity to see the score of similarity scores between films. The higher the score of cosine similarity then the film has a strong relationship/similarity and vice versa.
- Step 2: Getting Recomendations film that have good similarity on film and ranking this film based cosine similarity score.

### Cosine Similarity

**Cosine similarity** is a measure of similarity between two non-zero vectors of an inner product space that measures the cosine of the angle between them.

```python
##Import Libraries
from sklearn.metrics.pairwise import cosine_similarity
```

Get Cosine Similarity and output in List, why i create this because it will be used at a later step.

```python
cosine_sim=cosine_similarity(datafilm_index, datafilm_index)
cosine_sim
```

Based on the results of the cosine stated that the score between film cosine similiarity is very high around 0.99.

### Content Based Filtering Recomendations

Before create code content_based filtering. 
I am Reseting index 'Movie' where this dataset will be used in the next step.

```python
datafilm_index = datafilm_index.reset_index()
datafilm_index
```

Get Recomendations Film

```python
#Set feature 'Movie' to be title 
titles = datafilm_index['Movie']

#Set Index of feature 'Movie'
indices = pd.Series(datafilm_index.index, index=datafilm_index['Movie'])

#Content_Based Filtering
def get_recommendations(title):
    idx = indices[title]
    
    #using cosine_sim in list to get content_based filtering recomndations
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:51]
    movie_indices = [i[0] for i in sim_scores]
    return titles.iloc[movie_indices]
```

```python
get_recommendations('Gundala')
```

**Conclusion**:
- The result of the above code is Rekomndasi Film that has similarities with the film Gundala where it has been performed from the similarity score with the film ' Gundala '. So we can recommend the movie ' Spiderman: Far from Home ' to people who love the film Gundala.
